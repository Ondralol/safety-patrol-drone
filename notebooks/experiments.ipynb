# YOLOv5 Training from Scratch

```python
import os
from pathlib import Path
from ultralytics import YOLO

DATA_YAML = Path("../data/data.yaml").resolve()
assert DATA_YAML.exists(), f"data.yaml not found at {DATA_YAML}"
```

```python
model = YOLO("yolov5s.yaml")  # architecture only, no pretrained weights

results = model.train(
    data=str(DATA_YAML),
    epochs=3,
    imgsz=320,
    batch=16,
    project="../runs",
    name="scratch",
)
```

```python
val_metrics = model.val(split="val")
test_metrics = model.val(split="test")

print(f"Val  mAP@50:    {val_metrics.box.map50:.4f}")
print(f"Val  mAP@50-95: {val_metrics.box.map:.4f}")
print()
print(f"Test mAP@50:    {test_metrics.box.map50:.4f}")
print(f"Test mAP@50-95: {test_metrics.box.map:.4f}")
print()
print("Per-class (test mAP@50):")
for name, ap in zip(test_metrics.names.values(), test_metrics.box.ap50):
    print(f"  {name}: {ap:.4f}")
```
