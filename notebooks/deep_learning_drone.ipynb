```python
from ultralytics import YOLO
import shutil
shutil.copytree("/content/drive/MyDrive/uni/data", "/content/data")
DATA_YAML = "/content/data/data.yaml"

model = YOLO("yolov5s.yaml")

results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    pretrained = False,
    project="/content/runs",
    name="scratch",
)

# Evaluate
val_metrics = model.val(split="val")
test_metrics = model.val(split="test")

print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")

# Save
model.save("/content/drive/MyDrive/uni/models/yolov5s_scratch.pt")
```

```python
from ultralytics import YOLO

model = YOLO("yolov5s.pt")  # downloads pretrained COCO weights

results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    project="/content/runs",
    name="pretrained",
)

val_metrics = model.val(split="val")
test_metrics = model.val(split="test")

print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")

# Then copy best weights to Drive after:
import shutil
shutil.copy("/content/runs/pretrained/weights/best.pt",
            "/content/drive/MyDrive/uni/models/yolov5s_pretrained.pt")
```

```python
# YOLO11 from scratch:

model = YOLO("yolo11s.yaml")

results = model.train(
    data=DATA_YAML,
    imgsz=640,
    batch=16,
    pretrained=False,
    project="/content/runs",
    name="yolo11-scratch",
)

val_metrics = model.val(split="val")
test_metrics = model.val(split="test")
print(f"Val  mAP@50: {val_metrics.box.map50:.4f}")
print(f"Test mAP@50: {test_metrics.box.map50:.4f}")

shutil.copy("/content/runs/yolo11-scratch/weights/best.pt",
            "/content/drive/MyDrive/uni/models/yolo11_scratch.pt")
```
